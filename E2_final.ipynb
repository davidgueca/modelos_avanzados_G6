```python
# Importamos Librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_acf
```

Se importa la base MultiTimeline correspondiente a la seria de tiempo de la busqueda en Zoom, las cifras se encuentran están en millones.

```python
df = pd.read_csv('multiTimeline.csv', header=0, index_col=0)
```

```python
df.head()
```

Se toma una muestra de la base equivalente a los primeros 5 registos

```python
df.dtypes
```

```python
df.plot(figsize=(24,5), linewidth=5, fontsize=20)
```

¿Is the time series stationary?

RTA/= The series is not stationary because the graph shows that it has a trend and does not show a constant variation.


```python
df2 = df[['zoom']]
df2.rolling(12).mean().plot(figsize=(24,5), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

Construct and plot the trend component of the time series

RTA/ = We have removed most of the seasonality, the chart show only the trend

```python
df2.diff().plot(figsize=(24,5), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

Construct and plot the seasonal component of the time series

RTA/ = We have removed most of the trend, the chart show only the seasonal

```python
plt.figure(figsize=(24,5))
pd.plotting.autocorrelation_plot(df2);
```

the chart show a statistical significance of the correlation. In this case, you can say that the 'zoom' series is genuinely autocorrelated with a lag of 24 months.
